# NDC Data Cleaning
This data could be useful for data studies I'd like to do. Lets prep the data for further use in future projects.

First things first, lets load the data.

```python
from IPython.display import display, HTML
import pandas as pd

pack_df = pd.read_excel("./data/package")
```

Looks like relying on pandas isn't going to be good enough. Lets parse the files myself.

```python
def xlsParser(fileobj):
    text_lines = fileobj.readlines()
    columns = [x.strip() for x in text_lines[0].split("\t")]
    data = [x.strip().split('\t') for x in text_lines[1:]]
    df = pd.DataFrame(columns=columns, data=data)
    return df

with open("./data/package.xls","r") as package_file:
    pack_df = xlsParser(package_file)
    
print(pack_df.shape)
pack_df.head()
```

```python
with open("./data/product.xls","r") as product_file:
    prod_df = xlsParser(product_file)

print(prod_df.shape)
prod_df.head()
```

# Data Investigation

Lets see if there are duplicated entries, nulls etc. and try and understand the data here.

```python
for col in pack_df.columns:
    print(col.strip()," has ",pack_df[pack_df.duplicated(subset=col)].shape[0]," duplicated values.")
```
