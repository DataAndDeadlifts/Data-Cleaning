Looks like relying on pandas isn't going to be good enough. Lets parse the files myself.

```python
def xlsParser(fileobj):
    text_lines = fileobj.readlines()
    columns = [x.strip() for x in text_lines[0].split("\t")]
    data = [x.strip().split('\t') for x in text_lines[1:]]
    df = pd.DataFrame(columns=columns, data=data)
    return df

with open("./data/package.xls","r") as package_file:
    pack_df = xlsParser(package_file)
    
print(pack_df.shape)
pack_df.head()
```

# Whats next?
The product dataframe looks to contain information on formulations of drugs. What's interesting is the PRODUCTID column is the only column in both dataframes to have zero duplicated values. It appears to be the unique identifier for all drugs. But how will people query this dataset? I doubt people will look for the PRODUCTID, rather they would look for the Proprietary or Non-Proprietary name(s). So lets see if we can condense and group this dataset by proprietary name.

It appears that some duplicates are there because the producing company changed name, or the product changed hands in the market. Lets remove duplicate rows that contain the same drug but changed companies, selecting the most recent company to own the drug.

```python
def checkAllMatch(values,checkcols):
    """Given a set of dframe rows, return a count of the columns that contain the same values."""
    non_unique_dict = {}
    for col in [x for x in values.columns if x in checkcols]:
        if len(values[col].unique())<=1:
            continue
        else:
            return False
    return True

def companyChangeRemover(dataframe):
    """I'm looking for rows that are identical except for their company name.
    If there are such rows, remove the row with the older marketing approval date."""
    return_df = dataframe.copy()
    grouped_df = dataframe.groupby(["NONPROPRIETARYNAME"])
    duplicate_index = []
    ignore_cols = ['PRODUCTID','PRODUCTNDC','STARTMARKETINGDATE',"APPLICATIONNUMBER"]
    check_cols = [x for x in dataframe.columns if x not in ignore_cols]
    for group in grouped_df.groups:
        group_vals = grouped_df.get_group(group)
        if len(group_vals) > 1:
            if checkAllMatch(group_vals,check_cols):
                #If there is only one unique start marketing date, take the original NDC.
                if len(group_vals['STARTMARKETINGDATE'].unique())<=1:
                    #display(group_vals)
                    group_vals.sort_values('PRODUCTNDC',inplace=True)
                    drop_indices = group_vals.index[1:].values.tolist()
                    duplicate_index.extend(drop_indices)
                #Otherwise, sort by starting marketing date, keep the latest one. Drop the rest.
                else:
                    #display(group_vals)
                    group_vals.sort_values(['STARTMARKETINGDATE'],ascending=False,inplace=True)
                    drop_indices = group_vals.index[1:].values.tolist()
                    duplicate_index.extend(drop_indices)
    duplicate_index.sort()
    return_df.drop(duplicate_index,axis=0,inplace=True)
    return_df.reset_index(drop=True,inplace=True)
    return return_df
            
prod_df_nodupmarkets = companyChangeRemover(prod_df.iloc[:100])
prod_df_nodupmarkets.head()
```

Now there are no duplicate drug entries that were sold to new companies and re-listed!
