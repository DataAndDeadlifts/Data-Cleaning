Looks like relying on pandas isn't going to be good enough. Lets parse the files myself.

```python
def xlsParser(fileobj):
    text_lines = fileobj.readlines()
    columns = [x.strip() for x in text_lines[0].split("\t")]
    data = [x.strip().split('\t') for x in text_lines[1:]]
    df = pd.DataFrame(columns=columns, data=data)
    return df

with open("./data/package.xls","r") as package_file:
    pack_df = xlsParser(package_file)
    
print(pack_df.shape)
pack_df.head()
```

```python
with open("./data/product.xls","r") as product_file:
    prod_df = xlsParser(product_file)

print(prod_df.shape)
prod_df.head()
```

It appears that some duplicates are there because the producing company changed name, or the product changed hands in the market. Lets remove duplicate rows if that is the case and keep the row that was approved for marketing more recently.

```python
def checkAllMatch(values,checkcols):
    """Given a set of dframe rows, return a count of the columns that contain the same values."""
    non_unique_dict = {}
    for col in [x for x in values.columns if x in checkcols]:
        if len(values[col].unique())<=1:
            continue
        else:
            return False
    return True

def companyChangeRemover(dataframe):
    """I'm looking for rows that are identical except for their company name.
    If there are such rows, remove the row with the older marketing approval date."""
    return_df = dataframe.copy()
    grouped_df = dataframe.groupby(["NONPROPRIETARYNAME"])
    duplicate_index = []
    check_cols = [x for x in dataframe.columns if x not in ['PRODUCTID','PRODUCTNDC','STARTMARKETINGDATE']]
    for group in grouped_df.groups:
        group_vals = grouped_df.get_group(group)
        if len(group_vals) > 1:
            if checkAllMatch(group_vals,check_cols):
                if len(group_vals['STARTMARKETINGDATE'].unique())<=1:
                    group_vals.sort_values('PRODUCTNDC',inplace=True)
                    drop_indices = group_vals.index[1:]
                    duplicate_index.append(drop_indices)
                else:
                    group_vals.sort_values(['STARTMARKETINGDATE'],ascending=False,inplace=True)
                    display(group_vals)
                    drop_indices = group_vals.index[1:]
                    duplicate_index.append(drop_indices)
    return duplicate_index
            
companyChangeRemover(prod_df)
```
